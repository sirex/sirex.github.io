## Atitikties palyginimo rezultatai

Galiausiai, rankiniu būdu, apjungus Excel lenteles, gavau tokį rezultatą.


```python
%matplotlib inline

import jupyterthemes.jtplot as jtplot
from pandas_ods_reader import read_ods

jtplot.style(theme='monokai', fscale=1.5, figsize=(16, 8))

demand = read_ods(
    '/home/sirex/ivpk/poreikis/eim-apklausa-2021.ods',
    sheet='Poreikis',
    columns=[
        'poreikis',
        'tiekejas',
        'prasantysis',
        'rinkinys',
        'kada',
        'apimtis',
        'atitikimas',
    ]
)
demand = demand.iloc[1:]
demand['kada'].value_counts().sort_index(ascending=False).plot.barh();
```

Tačiau, šis rezultatas atspindi tik tai ar rinkinys konkrečiu pavadinimu yra
planuojamas atverti ar ne. Net jei tarkim yra deklaruojama, kad Nekilnojamojo
turto registro duomenys jau atverti, tai nereiškia, kad tai kas atverta pilnai
atitinka poreikį. Todėl nusprendžiau įtraukti dar vieną kriterijų, kuriame
bandžiau įvertinti ar tai, kas planuojama atverti, pilna apimtimi atitinka tai,
ko reikia.

```python
demand['atitikimas'].value_counts().plot.barh();
```

Iš čia, pilnai poreikį tenkina tai, kas pažymėta „Jau atverta“ arba
„Planuojama atverti“. Jei tiksliau, tai pilnai tenkinama 13 poreikių.

Lentelėje žemiau True yra rinkinių skaičius, kuris jau atvertas arba
planuojamas atverti ir pilnai tenkina poreikį, o False yra tai, kas nebus
atverta arba pilnai netenkina poreikio.

```python
atitinka = ['Jau atverta', 'Planuojama atverti']
demand['atitikimas'].isin(atitinka).value_counts().plot.barh()
demand['atitikimas'].isin(atitinka).value_counts()
```

Jei skaičiuoti dar tiksliau, tai poreikių lentelėje, dalis poreikių dubliuojasi,
pavyzdžiui Adresų registras buvo pažymėtas kelis kartus. Eliminavus dublikatus,
skaičiai būtų šiek tiek mažesni.
